```python
from pymongo import MongoClient
from pprint import pprint
from flask import Flask, jsonify,render_template
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# confirm that our new database was created
print(mongo.list_database_names())
```

```python
# assign the greenhouse_emission database to a variable name
greenhouse_emission = mongo['greenhouse_emission']
```

```python
# review the collections in our new database
pprint(greenhouse_emission.list_collection_names())
```

```python
# assign the collection to a variable
electricity_consumption = greenhouse_emission['electricity_consumption']
fugitive = greenhouse_emission['fugitive']
lulucf = greenhouse_emission['lulucf']
ippu = greenhouse_emission['ippu']
agriculture = greenhouse_emission['agriculture']
electricity_generation = greenhouse_emission['electricity_generation']
waste = greenhouse_emission['waste']
geo_data = greenhouse_emission['geo_data']
transport = greenhouse_emission['transport']
stationary_energy = greenhouse_emission['stationary_energy']
all_emissions = greenhouse_emission['all_emissions']
```

```python
print('electricity_consumption', electricity_consumption.count_documents({}))
print('fugitive', fugitive.count_documents({}))
print('lulucf', lulucf.count_documents({}))
print('ippu', ippu.count_documents({}))
print('agriculture', agriculture.count_documents({}))
print('electricity_generation', electricity_generation.count_documents({}))
print('waste', waste.count_documents({}))
print('transport', transport.count_documents({}))
print('stationary_energy', stationary_energy.count_documents({}))
print('geo_data', geo_data.count_documents({}))
print('all_emissions', all_emissions.count_documents({}))
```
