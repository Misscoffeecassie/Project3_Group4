```python
import pandas as pd
import json

# Read the CSV file
data = pd.read_csv('/Users/cassie/Documents/MONASH/Activity/Project3_Group4/Resources/all_data.csv')
# dispaly date
data.columns
 
Avg_Emissions_2016=data[data['Year']==2016].groupby("Planning_Region")["Emissions"].mean()
Avg_Emissions_2017=data[data['Year']==2017].groupby("Planning_Region")["Emissions"].mean()
Avg_Emissions_2018=data[data['Year']==2018].groupby("Planning_Region")["Emissions"].mean()
Avg_Emissions_2019=data[data['Year']==2019].groupby("Planning_Region")["Emissions"].mean()
Avg_Emissions_df=pd.DataFrame({'Year 2016':Avg_Emissions_2016,
                            'Year 2017':Avg_Emissions_2017,
                            'Year 2018':Avg_Emissions_2018,
                            'Year 2019':Avg_Emissions_2019}).reset_index()
Avg_Emissions_df

```

```python
# Specify the file path
file_path = 'Avg_Emissions.json'

# Save DataFrame as JSON
Avg_Emissions_df.to_json(file_path, orient='records')
```
